```python
import pandas as pd
```

# Imports

```python
# 2014 :
apr14 = pd.read_csv("uber-tlc-foil-response-master/uber-trip-data/uber-raw-data-apr14.csv")
may14 = pd.read_csv("uber-tlc-foil-response-master/uber-trip-data/uber-raw-data-may14.csv")
jun14 = pd.read_csv("uber-tlc-foil-response-master/uber-trip-data/uber-raw-data-jun14.csv")
jul14 = pd.read_csv("uber-tlc-foil-response-master/uber-trip-data/uber-raw-data-jul14.csv")
aug14 = pd.read_csv("uber-tlc-foil-response-master/uber-trip-data/uber-raw-data-aug14.csv")
sep14 = pd.read_csv("uber-tlc-foil-response-master/uber-trip-data/uber-raw-data-sep14.csv")
aprsep14 = pd.concat([apr14,may14,jun14,jul14,aug14,sep14], ignore_index=True)
```

```python
#2015
janjune15 = pd.read_csv("uber-tlc-foil-response-master/uber-trip-data/uber-raw-data-janjune-15.csv")
```

# Base information data

```python
d14 = pd.DataFrame({
    "Time":aprsep14["Date/Time"],
    "Base":aprsep14["Base"]
})
d15 = pd.DataFrame({
    "Time":janjune15["Pickup_date"],
    "Base":janjune15["Affiliated_base_num"]
})
all_data = pd.concat([d14,d15], ignore_index=True)
# all_data = d15 # Si on veut uniquement 2015
```

```python
all_data
```

```python
pickup_count = all_data["Base"].value_counts()
```

```python
base_to_keep = ["B02682","B02764","B02617","B02598","B02765","B02512"]
```

```python
all_data_reduced = all_data.query("Base in @base_to_keep")
```

```python
all_data_reduced["Time"] = pd.to_datetime(all_data_reduced["Time"])
```

```python
all_data_reduced_oh = pd.get_dummies(all_data_reduced)
```

# 2015 HOURLY

```python
data_hourly15 = all_data_reduced_oh.groupby([pd.Grouper(key='Time',freq='H')]).sum().reset_index()
data_hourly15
```

```python
data_hourly15.to_csv("data/hourly_main_bases_2015.csv")
```

# 2015 10 min

```python
data_10min15 = all_data_reduced_oh.groupby([pd.Grouper(key='Time',freq='10min')]).sum().reset_index()
data_10min15
```

```python
data_10min15.to_csv("data/hourly_main_bases_2015.csv")
```

# TOUT

```python
data_hourly = all_data_reduced_oh.groupby([pd.Grouper(key='Time',freq='H')]).sum().reset_index()
data_hourly
```

```python
data_hourly.to_csv("data/hourly_main_bases_all.csv")
```

# Location data

```python
d15_location = pd.DataFrame({
    "Time":janjune15["Pickup_date"],
    # "Location":janjune15["locationID"].to_string()
    "Location":janjune15["locationID"]
})

d15_location
```

```python
location_count = d15_location["Location"].value_counts()
location_count = location_count.sort_values(ascending=False).iloc[:20]
location_to_keep = location_count.index
```

```python
d15_location_reduced = d15_location.query("Location in @location_to_keep")
d15_location_reduced
```

```python
d15_location_reduced["Time"] = pd.to_datetime(d15_location_reduced["Time"])
```

```python
d15_location_oh = pd.get_dummies(d15_location_reduced, columns=["Location"])
```

## Par heure

```python
data_hourly = d15_location_oh.groupby([pd.Grouper(key='Time',freq='H')]).sum().reset_index()
data_hourly
```

```python
data_hourly.to_csv("data/hourly_main_locations_15.csv")
```

## 10 min

```python
data_10min = d15_location_oh.groupby([pd.Grouper(key='Time',freq='10min')]).sum().reset_index()
data_10min
```

```python
data_10min.to_csv("data/10min_main_locations_15.csv")
```
